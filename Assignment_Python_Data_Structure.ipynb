19. Why are dictionaries considered unordered, and how does that affect data retrieval?
 - Historically, dictionaries were considered "unordered" because their internal structure (the hash table) focused on where a key belongs in memory rather than when it was added.However, this definition changed in recent versions of Python.
 In older versions of Python, a dictionary was like a bag of marbles. When you added a key, Python calculated its hash and placed it in the first available "bucket" that matched that hash. memory effciency fast lookups.
 If you iterated through a dictionary, the items appeared in a seemingly random order. The order depended entirely on the hash values and the size of the memory table, not on your input.
  - this affects Data RetrievalRegardless of whether a dictionary is ordered or unordered, the method of retrieval remains the same,
  -  but the predictability changes:A. Retrieval by Key (The Primary Way)This is unaffected by order. Whether the dictionary is ordered or not, you still retrieve data by calling my_dict["key"]. Because this uses hashing, it is always $O(1)$ (instant), no matter where the key is "positioned.
  -Retrieval by Iteration (Loops)This is where the difference is felt.In older Python: If you looped through a dictionary to print names, they would come out in a random order every time the hash table resized.
  
  - In modern Python: If you loop through a dictionary, you are guaranteed to get the items back in the exact order you put them in. This is critical for tasks like generating a CSV header from dictionary keys.C. Retrieval by "Index" (Not possible)Even though modern dictionaries are ordered, they are still not indexable. You cannot ask for my_dict[0] to get the first item. You must still use the specific key name.
